```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Iris dataset
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
# Print the first five rows of the dataset
print(iris.head())
```

```python
# Print the shape of the dataset
print(iris.shape)
```

```python
# Print summary statistics of the dataset
print(iris.describe())
```

```python
# Plot the distribution of each feature
sns.set()
sns.histplot(data=iris, x="sepal_length", kde=True)
plt.show()

sns.histplot(data=iris, x="sepal_width", kde=True)
plt.show()

sns.histplot(data=iris, x="petal_length", kde=True)
plt.show()

sns.histplot(data=iris, x="petal_width", kde=True)
plt.show()
```

```python
# Plot pairwise scatterplots of the features
sns.pairplot(data=iris, hue="species")
plt.show()
```

```python
# Calculate the correlation matrix
corr = iris.corr()
```

```python
# Plot the correlation matrix as a heatmap
sns.heatmap(corr, cmap="YlGnBu", annot=True)
plt.show()
```
