```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the Titanic dataset
titanic_df = pd.read_csv('https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv')
```

```python
# View the first few rows of the dataset
print(titanic_df.head())
```

```python
# View the data types of each column
print(titanic_df.dtypes)
```

```python
# View the distribution of a numeric variable using a histogram
plt.hist(titanic_df['Age'].dropna(), bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Passengers Age')
plt.show()

```

```python
# View the distribution of a categorical variable using a bar chart
sns.countplot(x='Sex', data=titanic_df)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Distribution of Passengers by Sex')
plt.show()

```

```python
# View the relationship between two numeric variables using a scatterplot
sns.scatterplot(x='Age', y='Fare', data=titanic_df)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Relationship between Age and Fare')
plt.show()

```

```python
# View the relationship between a numeric variable and a categorical variable using a boxplot
sns.boxplot(x='Survived', y='Age', data=titanic_df)
plt.xlabel('Survived')
plt.ylabel('Age')
plt.title('Relationship between Survival and Age')
plt.show()
```

```python

# View the correlation between numeric variables using a heatmap
corr_matrix = titanic_df.corr()
sns.heatmap(corr_matrix,cmap="YlGnBu", annot=True)
plt.show()
```
